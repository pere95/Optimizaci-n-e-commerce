# 4.TRANSFORMACION DE DATOS

Vamos a crear 3 tipos de nuevas variables

* Extraer componentes
* Variables de calendario: Festivos locales (Rusia, ya que el ecommerce es ruso)
* Indicadores ex√≥genos: D√≠as no necesariamente festivos pero con inter√©s comercial: Black Friday, Cyber Monday, Reyes, San Valentin

```python
def comp_fecha(dataf):
    date = dataf.index.date
    a√±o = dataf.index.year
    mes = dataf.index.month
    dia = dataf.index.day
    hora = dataf.index.hour
    minuto = dataf.index.minute
    segundo = dataf.index.second

    return(pd.DataFrame({'date':date, 'a√±o':a√±o,'mes':mes, 'dia':dia, 'hora':hora, 'minuto':minuto, 'segundo':segundo}))
```

```python
df2=pd.concat([df.reset_index(),comp_fecha(df)],axis=1).set_index('fecha')
df2.isna().sum()
```

```python
import holidays
festivo_ru = holidays.RU(years=2020)
festivo_ru
```

```python
df2['festivo_ru']=df2.date.apply(lambda x:1 if (x in festivo_ru) else 0)
df2.festivo_ru.value_counts()
```

```python
df2['festivo']=np.where(df2.date.isin(festivo_ru),1,0)
df2.festivo.value_counts()
```

```python
#Comprobamos festivos
df2[df2.festivo == 1].date.value_counts().sort_index()
```

```python
#A√±ado black friday y san valentin
df2['black_friday'] = 0
df2.loc['2019-11-29','black_friday'] = 1

df2['san_valentin'] = 0
df2.loc['2020-02-14','san_valentin'] = 1
```

```python
df2.black_friday.value_counts()
```
